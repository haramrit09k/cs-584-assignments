# Experimenting locally to find best value of k and other hyperparameters

# Step 1: Data Preprocessing and Feature Selection

## 1.a Importing essential libraries and viewing the datasets

```python
import pandas as pd
import numpy as np
import re
import nltk
import string
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer, HashingVectorizer, TfidfVectorizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import statistics
from statistics import mode
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity

train_data = pd.read_csv("1598639150_466036_train_file.txt", sep="\t")
# print(train_data)

test_data = pd.read_csv("1598639150_4984343_test_file.txt", header=None, names=["review"])
# print(test_data)
```

## 1.b Ensuring test dataset has expected number of entries i.e. 15000 

```python
print("Before preprocessing the test set: size = ", test_data.size)
f = open("1598639150_4984343_test_file.txt", 'r', encoding="utf8")
txt = f.read().split('\n')
txt = list(filter(None, txt))
test_data = pd.DataFrame(txt, columns=['review'])
print("After preprocessing the test set: size = ", test_data.size)
```

## 1.c Check for null values in dataset

```python
print("Number of null values in train set:\n",train_data.isnull().sum())
```

## 1.d Dataset cleaning

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stopwords = stopwords.words('english')
stopwords.append('br') # to remove br tags 
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
# Convert everything to lower case
train_data['review'] = train_data['review'].str.lower()
# Remove all characters that are not alphabets
train_data['review'] = train_data['review'].str.replace('[^a-zA-Z ]', ' ')
# Lemmatization, removal of stopwords and only accepting words with length greater than equal to 2
train_data['review'] = train_data['review'].apply(lambda x : ' '.join([lemmatizer.lemmatize(word) for word in x.split() if word not in stopwords and len(word) > 1]))
```

```python
# Viewing first 10 elements of train dataset to confirm the cleaning process
print(train_data['review'][:10])
```

# Step 2: Finding the best value of k

## 2.a Split training set into train-test set (holdout) to test accuracy

```python
X_train, X_test, y_train, y_test = train_test_split(train_data['review'], train_data['sentiment'], test_size=0.2, random_state=42)
```

## 2.b Perform TFIDF vectorization of the datasets

```python
# initialize TFIDF vectorizer
vectorizer = TfidfVectorizer(min_df=2, ngram_range=(1,3), max_features = 75000)
# fit and transform the vectorizer on the training dataset
vectorizer.fit(X_train)
train_vector = vectorizer.transform(X_train)
# transform the vectorizer on the testing dataset
test_vector = vectorizer.transform(X_test)
```

```python
X_train = train_vector
X_test = test_vector
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# converted y_train to numpy array to resolve error I faced while finding the mode
y_train = y_train.to_numpy()
print(y_train.shape)
print(type(y_train))
```

```python
# converted y_train to numpy array to resolve a random error I faced while comparing predicted and true values
y_test = y_test.to_numpy()
print(y_test.shape)
print(type(y_test))
```

## 2.c Implementing the KNN algorithm

### (i) Calculating cosine similarity between X_train and X_test

```python
cos_sim = cosine_similarity(X_train, X_test)
cos_sim = cos_sim.transpose() #transpose operation to have the test set elements as rows for easier iteration
print(type(cos_sim))
print(cos_sim.shape)
print(cos_sim)
```

### (ii) Run the KNN algorithm for values of k in a set range to find the best value

```python
k_list = []
acc_list = []
```

```python
for k in range(1,1000,2):
    indices = list()
    for doc in cos_sim:
        indices.append(np.argsort(doc)[-k:])
    pred_val = list()
    for arr in indices:
        pred_val.append(mode([y_train[i] for i in arr]))
    correct = 0
    for i in range(len(pred_val)):
        if pred_val[i] == y_test[i]:
            correct+=1
    print("For k = ",k," => ", correct/len(pred_val)*100, "%\n")
    
    k_list.append(k)
    acc_list.append(correct/len(pred_val)*100)
```

```python
print(k_list)
```

### (iii) Finding the max value of k from the experiment values

```python
# Find index with max_accuracy
max_acc_index = acc_list.index(max(acc_list))
# Find value of k corresponding to the max index
selected_k = k_list[max_acc_index]
print("Selected value of k is",selected_k,"since it gave an accuracy of",acc_list[max_acc_index])
print(k_list[98])
np.argsort(acc_list)
```
