```python
def train_using_gini(X_train, X_test, y_train): 
  
    # Creating the classifier object 
    clf_gini = DecisionTreeClassifier(criterion = "gini",max_depth = 11) 
  
    # Performing training 
    clf_gini.fit(X_train, y_train) 
    return clf_gini
```

```python
def prediction(X_test, clf_object): 
  
    # Predicton on test with giniIndex 
    y_pred = clf_object.predict(X_test) 
    print("Predicted values:") 
    print(y_pred) 
    return y_pred 
```

```python
def cal_accuracy(y_test, y_pred): 
      
#     print("Confusion Matrix: ", 
#         confusion_matrix(y_test, y_pred)) 
      
    print ("Accuracy : ", 
    accuracy_score(y_test,y_pred)*100) 
      
#     print("Report : ", 
#     classification_report(y_test, y_pred)) 
    
    print ("F1 score : ", 
    f1_score(y_test,y_pred)*100) 
      
```

```python
piv = df.pivot_table(index=['userID'], columns=['movieID'], values='rating')
```

```python
piv.head(15)
```

```python
piv_norm = piv
piv_norm.fillna(0, inplace=True)
piv_norm = piv_norm.T
piv_norm = piv_norm.loc[:, (piv_norm != 0).any(axis=0)]
```

```python
piv_norm.head()
```

```python
piv_sparse = sp.sparse.csr_matrix(piv_norm.values)
```

```python
piv_sparse
```

```python
from sklearn.metrics.pairwise import cosine_similarity
item_similarity = cosine_similarity(piv_sparse)
```

```python
item_sim_df = pd.DataFrame(item_similarity, index = piv_norm.index, columns = piv_norm.index)
```

```python
def top_animes(anime_name):
    count = 1
    print('Similar shows to {} include:\n'.format(anime_name))
    for item in item_sim_df.sort_values(by = anime_name, ascending = False).index[1:11]:
        print('No. {}: {}'.format(count, item))
        count +=1  
```

```python
user_similarity = cosine_similarity(piv_sparse.T)
```

```python
user_sim_df = pd.DataFrame(user_similarity, index = piv_norm.columns, columns = piv_norm.columns)
```

```python
def top_users(user):
    
    if user not in piv_norm.columns:
        return('No data available on user {}'.format(user))
    
    print('Most Similar Users:\n')
    sim_values = user_sim_df.sort_values(by=user, ascending=False).loc[:,user].tolist()[1:11]
    sim_users = user_sim_df.sort_values(by=user, ascending=False).index[1:11]
    zipped = zip(sim_users, sim_values,)
    for user, sim in zipped:
        print('User #{0}, Similarity value: {1:.2f}'.format(user, sim)) 
```

```python
def similar_user_recs(user):
    
    if user not in piv_norm.columns:
        return('No data available on user {}'.format(user))
    
    sim_users = user_sim_df.sort_values(by=user, ascending=False).index[1:11]
    best = []
    most_common = {}
    
    for i in sim_users:
        max_score = piv_norm.loc[:, i].max()
        best.append(piv_norm[piv_norm.loc[:, i]==max_score].index.tolist())
    for i in range(len(best)):
        for j in best[i]:
            if j in most_common:
                most_common[j] += 1
            else:
                most_common[j] = 1
    sorted_list = sorted(most_common.items(), key=operator.itemgetter(1), reverse=True)
    return sorted_list[:5]    
```

```python
def predicted_rating(anime_name, user):
    if user not in piv_norm.columns or anime_name not in piv_norm.T.columns:
        return 2.5
    sim_users = user_sim_df.sort_values(by=user, ascending=False).index[1:10]
    user_values = user_sim_df.sort_values(by=user, ascending=False).loc[:,user].tolist()[1:10]
    rating_list = []
    weight_list = []
    for j, i in enumerate(sim_users):
        rating = piv.loc[i, anime_name]
        if not rating:
            return 2.5
        similarity = user_values[j]
        if np.isnan(rating):
            continue
        elif not np.isnan(rating):
            rating_list.append(rating*similarity)
            weight_list.append(similarity)
    if sum(rating_list)!=0 and sum(weight_list)!=0:
        return sum(rating_list)/sum(weight_list) 
    else:
        return 2.5
```
